```python
import matplotlib.pyplot as plt

months = ['sep', 'oct', 'nov', 'dec', 'jan', 'feb', 'march']
my_weight = [40, 41.5, 40, 42, 43, 42, 45]  # Converted to floats
sister_weight = [60, 62, 59, 58, 60, 58, 56]  # Corrected list definition

plt.plot(months, my_weight, color='red', label='My Weight')
plt.plot(months, sister_weight, color='blue', label='Sister Weight')

plt.xlabel('Months')
plt.ylabel('Weight in kgs')

plt.title('Weight Tracker for 7 Months')
plt.legend()

plt.show()
```
